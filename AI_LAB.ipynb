```python

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
group1 = [10, 15, 20, 25]
group2 = [5, 10, 15, 20]



plt.bar(categories, group1, color='blue', label='Group 1')
plt.bar(categories, group2, color='orange', bottom=group1, label='Group 2')



plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
group1 = [10, 15, 20, 25]
group2 = [5, 10, 15, 20]
group3 = [15, 10, 5, 25]

# Group 1 is at the bottom
plt.bar(categories, group1, color='blue', label='Group 1')

# Group 2 is stacked on top of Group 1
plt.bar(categories, group2, color='orange', bottom=group1, label='Group 2')

# Group 3 is stacked on top of (Group 1 + Group 2)
bottom_values = np.array(group1) + np.array(group2)
plt.bar(categories, group3, color='black', bottom=bottom_values, label='Group 3')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

categories=['A','B','C','D']
values=[10,15,20,25];

plt.bar(categories,values,color=['blue','red','green','yellow'])

plt.xlabel='categories'
plt.ylabel='values'
plt.title='bar chart'
plt.show()
```

```python
import matplotlib.pyplot as plt

labels =['Apple','Banana','Cherry','Date']
sizes=[30,20,40,10]
colors=['red','yellow','pink','brown']

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=140)

plt.title('pie chart Example')

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df=sns.load_dataset("tips")

sns.regplot(x="total_bill",y="tip",data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.title("regration plot")
plt.show()
```
