```python
#Bar Chart using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

# Create bar chart
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Display the plot
plt.show()
```

```python
# Stacked Bar Chart using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
group1 = [10, 15, 20, 25]
group2 = [5, 10, 15, 20]

# Create stacked bar chart
plt.bar(categories, group1, color='blue', label='Group 1')
plt.bar(categories, group2, color='orange', bottom=group1, label='Group 2')

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
#Pie Chart using Matplotlib
import matplotlib.pyplot as plt

# Data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 20, 40, 10]
colors = ['red', 'yellow', 'pink', 'brown']

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

```python
#Violin Plot using Seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Create violin plot
sns.violinplot(x="day", y="total_bill", data=df, palette="coolwarm")

# Show plot
plt.title('Violin Plot Example')
plt.show()
```

```python
#KDE (Kernel Density Estimation) Plot using Seaborn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


# Generate random data
data = np.random.randn(1000)

# Create KDE plot
sns.kdeplot(data, shade=True, color="blue")

# Show plot
plt.title('KDE Plot Example')
plt.show()
```

```python
#Swarm Plot using Seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Create swarm plot
sns.swarmplot(x="day", y="total_bill", data=df, palette="coolwarm")

# Show plot
plt.title('Swarm Plot Example')
plt.show()
```

```python
#Regression Plot using Seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Create regression plot
sns.regplot(x="total_bill", y="tip", data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})

# Show plot
plt.title('Regression Plot Example')
plt.show()
```

```python
#Multiple Line Plot using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot data
plt.plot(x, y1)
plt.plot(x, y2)

# Labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line Plot')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
