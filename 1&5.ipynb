```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

categories=['A','B','C','D']
sizes=[23,56,96,36]

plt.bar(categories,sizes,color=['blue','green','red','yellow'])

plt.title('Bar Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

categories=['A','B','C','D']
group1=[12,36,56,98]
group2=[15,25,63,96]

plt.bar(categories,group1,color='blue',label='group1')
plt.bar(categories,group2,color='green',bottom='group1',label='group2')

plt.title("Group barchart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data=np.random.randn(1000)
sns.kdeplot(data,shade=True,color='red')

plt.title("KDE Plot")
plt.show()








```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns



df=sns.load_dataset("tips")
sns.swarmplot(x="day",y="total_bill",data=df,palette="coolwarm")

plt.title("Swarm plot")
plt.show()



```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df=sns.load_dataset("tips")

sns.regplot(x="total_bill",y="tip",data=df,scatter_kws={"color":"red"},line_kws={"color":"green"})

plt.title("Regration plot")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


labels=['Apple','Angur','komola','malta']
colors=['blue','green','red','black']
sizes=[36,45,89,63]

plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",startangle=140)

plt.title("pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label='sine wave',color='blue')
plt.plot(x,y2,label='cosine wave',color='green')


plt.legend()
plt.show()
```
