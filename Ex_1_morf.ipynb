```python
# -------------------------------
# 1. Load Dataset
# -------------------------------
from sklearn.datasets import load_iris
import pandas as pd


data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

#for custom dataset

#df = pd.read_csv("data.csv")
#X = df.iloc[:, :-1]
#y = df.iloc[:, -1]

print("Dataset Loaded Successfully!")
print("X shape:", X.shape)
print("y shape:", y.shape)


# -------------------------------
# 2. Preprocessing
# -------------------------------
# Check missing values
print("\nMissing Values:")
print(X.isnull().sum())

# Standard Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nData Scaled Successfully!")


# -------------------------------
# 3. Visualization (Scatter Plot)
# -------------------------------
import matplotlib.pyplot as plt

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y)
plt.xlabel("Feature 1: " + X.columns[0])
plt.ylabel("Feature 2: " + X.columns[1])
plt.title("Scatter Plot of Iris Dataset")
plt.show()


# -------------------------------
# 4. Train-Test Split
# -------------------------------
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

print("\nTrain-Test Split Done!")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)


# -------------------------------
# 5. ML Model (Logistic Regression)
# -------------------------------
#from sklearn.linear_model import LogisticRegression
#model = LogisticRegression()
#model.fit(X_train, y_train)


# 5. ML Model (Kmeans)
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)   # 3 cluster example
model.fit(X_scaled)



# 5. ML Model (Decision tree classifier)
#from sklearn.tree import DecisionTreeClassifier
#model = DecisionTreeClassifier()
#model.fit(X_train, y_train)


# 5. ML Model (Knn)
#from sklearn.neighbors import KNeighborsClassifier
#model = KNeighborsClassifier(n_neighbors=5)   # 5 neighbors example
#model.fit(X_train, y_train)

print("\nModel Training Completed!")
# -------------------------------
# 6. Model Evaluation
# -------------------------------
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("\nAccuracy:", acc)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
